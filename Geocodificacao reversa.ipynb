```python
import re
import pandas as pd
#import shapely.geometry
#import geopy
from geopy.geocoders import Nominatim
import geopandas
from shapely import wkt
```

```python
end = pd.read_csv('.\aldeias_pontos.csv', encoding = 'UTF-8', sep = ';')
```

```python
end.head()
```

```python
end['the_geom'] = geopandas.GeoSeries.from_wkt(end['the_geom'])
end["coord_lat"] = end["the_geom"].map(lambda p: p.y)
end["coord_long"] = end["the_geom"].map(lambda p: p.x)
end["lat_long"] = end["the_geom"].map(lambda p: str(p.y) + ", " + str(p.x))

end.head()
```

```python
locator = Nominatim(user_agent="myGeocoder", timeout=10)
coordinates = "-11.3278444444, -53.2047777778"
location = locator.reverse(coordinates)
location.raw
```

```python
locator = Nominatim(user_agent="myGeocoder", timeout=10)
for index, row in end.iterrows():
    coordinates = row.lat_long
    location = locator.reverse(coordinates)
    end.loc[index, 'endereco'] = location.raw['display_name']
    for key, value in location.raw['address'].items():
        if key == 'postcode':
            end.loc[index, 'cep'] = location.raw['address']['postcode']
    
```
